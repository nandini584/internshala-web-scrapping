```python
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

url = 'https://www.ambitionbox.com/list-of-companies?page=1'

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # This will raise an HTTPError if the HTTP request returned an unsuccessful status code
    print(response.text)
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
webpage=requests.get('https://internshala.com/internships/').text
```

```python
soup=BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
for i in range(0, len(soup.find_all('h3'))):
  print(soup.find_all('h3')[i].text.strip())
```

```python
for i in soup.find_all('p', class_='company-name'):
  print(i.text.strip())
```

```python
for i in soup.find_all('span', class_='stipend'):
  print(i.text.strip())
```

```python
internship=soup.find_all('div', class_="individual_internship")
```

```python
name=[]
company=[]
time=[]
stipend=[]
location=[]
for i in internship:
  if i.find('h3') and i.find('p', class_='company-name') and i.find_all('span')[0] and i.find_all('span')[1] and i.find('span', class_="stipend"):
    name.append(i.find('h3').text.strip())
    company.append(i.find('p', class_='company-name').text.strip())
    location.append(i.find_all('span')[0].text.strip())
    time.append(i.find_all('span')[1].text.strip())
    stipend.append(i.find('span', class_="stipend").text.strip())

d={'name':name, 'company':company, 'time': time, 'stipend':stipend, 'location': location}
df=pd.DataFrame(d)
df
```

```python
final=pd.DataFrame()
for j in range(1,11):
  webpage=requests.get('https://internshala.com/internships/page-{}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  internship=soup.find_all('div', class_="individual_internship")
  name=[]
  company=[]
  time=[]
  stipend=[]
  location=[]
  for i in internship:
    if i.find('h3') and i.find('p', class_='company-name') and i.find_all('span')[0] and i.find_all('span')[1] and i.find('span', class_="stipend"):
      name.append(i.find('h3').text.strip())
      company.append(i.find('p', class_='company-name').text.strip())
      location.append(i.find_all('span')[0].text.strip())
      time.append(i.find_all('span')[1].text.strip())
      stipend.append(i.find('span', class_="stipend").text.strip())

  d={'name':name, 'company':company, 'time': time, 'stipend':stipend, 'location': location}
  df=pd.DataFrame(d)
  final = pd.concat([final, df], ignore_index=True)
```
